```python
import pandas as pd
```

```python
books = pd.read_csv("books.csv")
keywords = pd.read_csv("book_keywords.csv")
```

```python
books.info()
```

```python
keywords.info()
```

```python
# Merge both sources on book_id (mirrors merging movies + credits)
df = pd.merge(books, keywords, on="book_id", how="inner")
```

```python
df.info()
```

```python
# Build a single 'tags' column that combines genre, keywords, author and description.
# This is the text the recommender will compare between books (same idea as the
# movie project's tags column built from genres, keywords, cast and crew).
df['author_clean'] = df['author'].str.replace(' ', '-')
df['genre_clean'] = df['genre'].str.replace(' ', '-')
df['tags'] = df['genre_clean'] + ',' + df['keywords'] + ',' + df['author_clean'] + ',' + df['description']
```

```python
df = df[['book_id', 'title', 'tags']]
df.rename(columns={'book_id': 'id'}, inplace=True)
```

```python
df.info()
```

```python
df['tags'] = df['tags'].str.lower()
```

```python
df.to_csv('cleaned_books.csv', index=False)
```
